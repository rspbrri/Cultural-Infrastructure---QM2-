```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

###Libraries

```python
libraries = pd.read_csv("gdrive/My Drive/data/Libraries.csv")
libraries.head()
```

```python
#discarding unnecessary data
libraries = libraries[['name',
                       'borough_name',
                       'borough_code',
                       'longitude',
                       'latitude']]
```

###Theatres

```python
theatres = pd.read_csv("gdrive/My Drive/data/Theatres.csv")
theatre_rehearsal = pd.read_csv("gdrive/My Drive/data/Theatre_rehearsal_studio.csv")
#discarding unnecessary data
theatres = theatres[['name',
                           'borough_name',
                     'borough_code',
                           'longitude',
                           'latitude']]
theatre_rehearsal = theatre_rehearsal [['name',
                                'borough_name',
                                'borough_code',
                                'longitude',
                                'latitude']] 
# concatenating the two theatre-related data frames 
t_all = pd.concat([theatres, theatre_rehearsal])
t_all.reset_index(drop=True)
# removing dublicate entries
t_all.drop_duplicates(subset='name',
                      keep='first',
                      inplace=True)
```

###Museums and Galleries

```python
musgal = pd.read_csv("gdrive/My Drive/data/Museums_and_public_galleries.csv")
#discarding unnecessary data
musgal = musgal[['name',
                 'borough_name',
                       'borough_code',
                       'longitude',
                       'latitude']]
musgal.head()
```

###Music Venues

```python
music = pd.read_csv("gdrive/My Drive/data/Music_venues_all.csv")
music_reh = pd.read_csv("gdrive/My Drive/data/Music_rehearsal_studios.csv")
#discarding unnecessary data
music = music[['name',
                       'borough_name',
                      'borough_code',
                       'longitude',
                       'latitude']]
music_reh = music_reh[['name',
                       'borough_name',
                      'borough_code',
                       'longitude',
                       'latitude']]
music.head()
```

###Dance Venues

```python
d_rehearsal = pd.read_csv("gdrive/My Drive/data/Dance_rehearsal_studios.csv")
d_performance = pd.read_csv("gdrive/My Drive/data/Dance_performance_venues.csv")

#discarding unnecessary data
d_rehearsal = d_rehearsal[['name',
                           'borough_name',
                           'borough_code',
                           'longitude',
                           'latitude']]
d_performance = d_performance [['name',
                                'borough_name',
                                'borough_code',
                                'longitude',
                                'latitude']]                          
```

```python
# concatenating the two dance-related data frames 
d_all = pd.concat([d_rehearsal, d_performance])
d_all.reset_index(drop=True)
# removing dublicate entries
d_all.drop_duplicates(subset='name',
                      keep='first',
                      inplace=True)
```

##All datasets together


```python
# include arts centre data in all (venues that provide dance and music and theatre)
arts_centres = pd.read_csv("gdrive/My Drive/data/Arts_centres.csv")
```

```python
all = pd.concat([d_all, libraries, t_all, music, musgal, arts_centres])
all.reset_index(drop=True)
```

```python
# removing duplicate entries after concatenating
all.drop_duplicates(subset='name',
                      keep='first',
                      inplace=True)
```

##Exporting datasets

```python
d_all.to_csv('dance_venues.csv')
libraries.to_csv('library_venues.csv')
t_all.to_csv('theatre_venues.csv')
music.to_csv('music_venues.csv')
musgal.to_csv('museum_gallery_venues.csv')
all.to_csv('all_venues.csv')
```

```python
import matplotlib.pyplot as plt
xdata = ['above $35,000', 'below $ 35,000']
ydata = [48,69]
plt.bar(xdata, ydata)
plt.ylim(ymin=0) 
plt.ylim(ymax=100) 
plt.title('Non participation in Arts & Museum by Income')
plt.xlabel('Income (US Dollars)')
plt.ylabel('Non-participation (Arts & Museum )(%)')
plt.show()
```
